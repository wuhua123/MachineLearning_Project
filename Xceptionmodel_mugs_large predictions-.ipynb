```python
import matplotlib.pyplot as plt
%matplotlib inline

import os
import numpy as np
import tensorflow as tf

from tensorflow import keras

from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.xception import (
    Xception, preprocess_input, decode_predictions)
```

```python
# Load the Xception model
# https://keras.io/applications/#xception
model = Xception(
    include_top=True,
    weights='imagenet')
# Default Image Size for Xception
image_size = (299, 299)
```

```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir("./pic") if isfile(join("./pic", f))]
onlyfiles

predictions_column = []
results_column = []


for file in onlyfiles:
    image_path = os.path.join("pic",file)
    img = image.load_img(image_path, target_size=image_size)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    predictions = model.predict(x)
#     for result in decode_predictions(predictions)[0]:
#       print(result[1])
    predictions_column.append(decode_predictions(predictions)[0][0][1])
    results_column.append(decode_predictions(predictions)[0][0][2])
                        
#     print('Predicted:', decode_predictions(predictions, top=1))
#     plt.imshow(img)
#     print(file)


```

```python
import pandas as pd
import scipy.io
import numpy as np
df_img = pd.DataFrame({"Predictions":predictions_column,
                       "Result":results_column,
                       
})
```
