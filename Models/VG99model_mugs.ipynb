```python
model = VGG19(include_top=True, weights='imagenet')
image_size = (224, 224)
```

```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir("./images") if isfile(join("./images", f))]
onlyfiles

predictions_column = []
results_column = []


for file in onlyfiles:
    image_path = os.path.join("images",file)
    img = image.load_img(image_path, target_size=image_size)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    predictions = model.predict(x)
#     for result in decode_predictions(predictions)[0]:
#       print(result[1])
    predictions_column.append(decode_predictions(predictions)[0][0][1])
    results_column.append(decode_predictions(predictions)[0][0][2])
                        
#     print('Predicted:', decode_predictions(predictions, top=1))
#     plt.imshow(img)
#     print(file)


```

```python
df_img = pd.DataFrame({"Predictions":predictions_column,
                       "Result":results_column,
                       "Actual": ""
})
```
